```python
#gamma correction
def adjust_gamma(image, gamma=1.0):
    # build a lookup table mapping the pixel values [0, 255] to
    # their adjusted gamma values
    invGamma = 1.0 / gamma
    table = np.array([((i / 255.0) ** invGamma) * 255
        for i in np.arange(0, 256)]).astype("uint8")
    # apply gamma correction using the lookup table
    return cv2.LUT(image, table)


def minirotate(img, label, path,clas):
    colorImage  = Image.open(img)
    for i in range(6):
        rotated     = colorImage.rotate(i)
        rotated = rotated.save("temp.jpg")
        img = cv2.imread("temp.jpg")
        h,w,d =img.shape
        d = i*5
        crop_img = img[d:h-d, d:w-d]
        resized = cv2.resize(crop_img, (224, 224), interpolation = cv2.INTER_AREA) 
        cv2.imwrite(path + '/' +str(clas)+ '/' + label + 'r{}.jpg'.format(i), resized)
        
    
# create gamma images & save them
def create_gamma(img,label,path,clas):
#     gam_025 = adjust_gamma(img,0.25)
#     cv2.imwrite(path + '/' + label + 'g1.jpg',gam_025)
    gam_2 = adjust_gamma(img,2)
    gam_2 = cv2.resize(gam_2, (224, 224), interpolation = cv2.INTER_AREA) 
    cv2.imwrite(path + '/' +str(clas)+ '/' + label + 'g2.jpg',gam_2)
    
def creat_affine(img,label,path,clas):
    img_flip_X = cv2.flip(img, 0)
#     print(label)
#     cv2.imwrite(path + '/' + label + 'a1.jpg',img_flip_X)
#     img_flip_XY = cv2.flip(img, -1)
#     cv2.imwrite(path + '/' + label + 'a2.jpg',img_flip_XY)
    img_flip_Y = cv2.flip(img, 11)
    img_flip_Y = cv2.resize(img_flip_Y, (224, 224), interpolation = cv2.INTER_AREA) 
    cv2.imwrite(path + '/' +str(clas)+ '/' + label + 'a3.jpg',img_flip_Y)

# create affine images & save them in new directory for all training inputs
for i in range(len(dirs)) : 
    name = dirs.loc[i, "file_name"][:-4]
    clas = dirs.loc[i, "annotation"]
    img = name + ".jpg"
    img_path = os.path.join(paths, img)
    img_label = name
    img_ = cv2.imread(img_path)
    creat_affine(img_,img_label,pathd,clas)
    create_gamma(img_,img_label,pathd, clas)
    minirotate(img_path,img_label,pathd, clas)
```
